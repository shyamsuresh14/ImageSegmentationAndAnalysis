```python
#max pixel value of the 200 slices

m = 0
for i in range(200):
    view = sitk.GetArrayViewFromImage(img)[i, :, :]
    if max(np.unique(view)) > m:
        m = max(np.unique(view))
m 
```

```python
def convertToGrayScaleRange(image):
    #m = max(np.unique(image))
    #m = 3008
    new_img = np.array(image)
    new_img[:, :] = (image/m)*255
    return new_img
```

```python
#Folder of images

for i in range(200):
    view = sitk.GetArrayViewFromImage(img)[i, :, :]
    grayView = view
    #grayView = convertToGrayScaleRange(view)
    ind = str(i+1)
    if len(ind) < 3:
        ind = "0"*(3 - len(ind)) + ind
    #cv2.imwrite("imageSlices_001/slice" + ind + ".png", grayView)
    cv2.imwrite("D:\\dataset_seg\\ImageSegmentationAndAnalysis\\4_Dataset_Seg\\imageMasks_001\\mask" + ind + ".png", grayView)
```

```python
#Preparing train dataset 200x15

for i in range(1, 16):
    imgName = "image00" + str(i) if len(str(i)) == 1 else "image0" + str(i)
    print(imgName, end=" , ")
    img_i = sitk.ReadImage('D:\\dataset_seg\\xVertSeg.v1\\Data1\\images\\' + imgName + '.mhd')
    no = 0
    
    try:
        m = 0
        for j in range(200):
            view = sitk.GetArrayViewFromImage(img_i)[j, :, :]
            if max(np.unique(view)) > m:
                m = max(np.unique(view))
            no += 1    

        for k in range(200):
            view = sitk.GetArrayViewFromImage(img_i)[k, :, :]
            grayView = convertToGrayScaleRange(view)
            ind = str(k+1)
            if len(ind) < 3:
                ind = "0"*(3 - len(ind)) + ind
            cv2.imwrite("D:\\dataset_seg\\train\\" + imgName + "_" + ind + ".png", grayView)
            
    except Exception as e:
        print(e, end=" , ")
        shape = 0
        for k in range(no):
            view = sitk.GetArrayViewFromImage(img_i)[k, :, :]
            grayView = convertToGrayScaleRange(view)
            ind = str(k+1)
            if len(ind) < 3:
                ind = "0"*(3 - len(ind)) + ind
            cv2.imwrite("D:\\dataset_seg\\train\\" + imgName + "_" + ind + ".png", grayView)
            shape = grayView.shape
        
    print(no, "images")
```

```python
#hdf5 to tiff

import h5py
hdf = h5py.File('D:\dataset_seg\exports\im2_comp.h5', 'r')
data = hdf['exported_data']
data
```

```python
#Convert prob map to B&W
def conv(x):    
    a = x[:, :, 0]
    b = x[:, :, 1]
    a[a > 0.5] = True
    a[a <= 0.5] = False
    return a
#plt.imshow(a, cmap="gray")
```

```python
#Convert all slices
newData = []
for i in range(200):
    newData.append(conv(data[i]))
```

```python
newData = np.array(newData)
newData.shape
```

```python
newImg = sitk.GetImageFromArray(newData)
newImg
```

```python
sitk.WriteImage(newImg, "D:\\dataset_seg\\output\\out.tif")
```
